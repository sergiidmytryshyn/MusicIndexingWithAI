```python
#!/usr/bin/env python3
import json
import os
import time
from collections import defaultdict
from pathlib import Path

import requests
from dotenv import load_dotenv

load_dotenv()

MB_USER_AGENT = os.getenv("MB_USER_AGENT", "nlp-project/1.0 (you@example.com)")
MUSICBRAINZ_URL = "https://musicbrainz.org/ws/2/artist"
REQUEST_SLEEP = 1.2  # щоб не ловити rate limit

# ---------------------------
# ТУТ ПОВИНЕН БУТИ СПИСОК АРТИСТІВ
# ---------------------------
# приклад:
# artist_names = ["Wu-Tang Clan", "Metallica", "Nirvana"]
artist_names: list[str] = all_artists_balanced  # <- ЗАПОВНИ СВОЇМ СПИСКОМ


# ---------------------------------
# 1) Мапи для країни та регіону
# ---------------------------------

COUNTRY_CODE_TO_NAME = {
    "FR": "France",
    "GB": "United Kingdom",
    "DE": "Germany",
    "ES": "Spain",
    "IE": "Ireland",
    "HR": "Croatia",
    "BE": "Belgium",

    "US": "United States",
    "CA": "Canada",
    "PR": "Puerto Rico",

    "BR": "Brasil",
    "AR": "Argentina",
    "CO": "Columbia",

    "ID": "Indonesia",
    "KR": "South Korea",

    "AU": "Australia",
    "NZ": "New Zealand",
}

COUNTRY_CODE_TO_REGION = {
    # Europe
    "FR": "Europe",
    "GB": "Europe",
    "DE": "Europe",
    "ES": "Europe",
    "IE": "Europe",
    "HR": "Europe",
    "BE": "Europe",

    # North America
    "US": "North America",
    "CA": "North America",
    "PR": "North America",

    # South America
    "BR": "South America",
    "AR": "South America",
    "CO": "South America",

    # Asia
    "ID": "Asia",
    "KR": "Asia",

    # Oceania
    "AU": "Oceania",
    "NZ": "Oceania",
}


# ---------------------------------
# 2) MusicBrainz API
# ---------------------------------

def get_artist_location(artist_name: str):
    params = {
        "query": f'artist:"{artist_name}"',
        "fmt": "json",
        "limit": 1,
    }
    headers = {"User-Agent": MB_USER_AGENT}

    try:
        resp = requests.get(MUSICBRAINZ_URL, params=params, headers=headers, timeout=15)
        resp.raise_for_status()
        data = resp.json()
    except Exception:
        return None, None

    artists = data.get("artists") or []
    if not artists:
        return None, None

    a = artists[0]

    country_code = a.get("country")

    city = None
    begin_area = a.get("begin-area")
    if isinstance(begin_area, dict):
        city = begin_area.get("name")

    # якщо нема begin-area, іноді місто/регіон в area.name
    if not city:
        area = a.get("area")
        if isinstance(area, dict):
            city = area.get("name")

    return country_code, city


# ---------------------------------
# 3) Стартова структура JSON
# ---------------------------------

def init_world_structure():
    return {
        "Parts_of_the_World": [
            {
                "name": "Europe",
                "countries": [
                    {"name": "France", "cities": ["Paris", "Nancy"]},
                    {"name": "United Kingdom", "cities": ["London", "England"]},
                    {"name": "Germany", "cities": []},
                    {"name": "Spain", "cities": []},
                    {"name": "Ireland", "cities": []},
                    {"name": "Croatia", "cities": []},
                    {"name": "Belgium", "cities": []},
                ],
            },
            {
                "name": "North America",
                "countries": [
                    {
                        "name": "United States",
                        "cities": ["Los Angeles", "Hawaii", "Illinois", "Miami"],
                    },
                    {"name": "Canada", "cities": []},
                    {"name": "Puerto Rico", "cities": []},
                ],
            },
            {
                "name": "South America",
                "countries": [
                    {"name": "Brasil", "cities": ["Sao Paulo"]},
                    {"name": "Argentina", "cities": ["Buenos Aires"]},
                    {"name": "Columbia", "cities": []},
                ],
            },
            {
                "name": "Asia",
                "countries": [
                    {"name": "Indonesia", "cities": []},
                    {"name": "South Korea", "cities": []},
                ],
            },
            {
                "name": "Oceania",
                "countries": [
                    {"name": "Australia", "cities": []},
                    {"name": "New Zealand", "cities": []},
                ],
            },
            {
                "name": "unknown",
                "cities": [],
            },
        ]
    }


def get_region_block(world_json: dict, region_name: str) -> dict | None:
    for part in world_json["Parts_of_the_World"]:
        if part["name"] == region_name:
            return part
    return None


def add_city_to_world(world_json: dict, region: str, country_name: str | None, city: str | None):
    # unknown: кладемо просто в список cities
    if region == "unknown":
        unknown_block = get_region_block(world_json, "unknown")
        if unknown_block is None:
            return
        if city and city not in unknown_block["cities"]:
            unknown_block["cities"].append(city)
        return

    region_block = get_region_block(world_json, region)
    if region_block is None:
        return

    if "countries" not in region_block:
        region_block["countries"] = []

    # шукаємо країну
    countries = region_block["countries"]
    country_obj = None
    if country_name:
        for c in countries:
            if c["name"] == country_name:
                country_obj = c
                break

        if country_obj is None:
            country_obj = {"name": country_name, "cities": []}
            countries.append(country_obj)

        if city and city not in country_obj["cities"]:
            country_obj["cities"].append(city)


# ---------------------------------
# 4) Основна логіка
# ---------------------------------

def main(output_path: str = "artists_locations.json"):
    if not artist_names:
        raise ValueError("artist_names порожній. Заповни список артистів перед запуском.")

    world = init_world_structure()

    for artist in artist_names:
        country_code, city = get_artist_location(artist)

        if not country_code:
            # нічого не знайшли — кидаємо в unknown, як "місто" = ім'я артиста
            add_city_to_world(world, "unknown", None, artist)
        else:
            region = COUNTRY_CODE_TO_REGION.get(country_code, "unknown")
            country_name = COUNTRY_CODE_TO_NAME.get(country_code, country_code)
            add_city_to_world(world, region, country_name, city or artist)

        time.sleep(REQUEST_SLEEP)

    with open(output_path, "w", encoding="utf-8") as f:
        json.dump(world, f, ensure_ascii=False, indent=2)

    print(f"✅ Saved locations JSON to {output_path}")


if __name__ == "__main__":
    main()
```

```python
import json
from pathlib import Path

# ==== ШЛЯХИ ДО ФАЙЛІВ ====
ULTRA_PATH = Path("../data/ultra_balanced_merged_cleaned.json")
LOCATIONS_PATH = Path("artists_locations.json")
OUTPUT_PATH = Path("artists_locations_updated.json")  # куди зберегти оновлений

# ==== МАПА КРАЇНА -> ЧАСТИНА СВІТУ ====
COUNTRY_TO_REGION = {
    # Europe
    "France": "Europe",
    "United Kingdom": "Europe",
    "Germany": "Europe",
    "Ireland": "Europe",
    "Austria": "Europe",
    "Norway": "Europe",
    "Sweden": "Europe",
    "Ukraine": "Europe",
    "Croatia": "Europe",
    "Belgium": "Europe",
    "Spain": "Europe",
    "Poland": "Europe",
    "Italy": "Europe",
    "Netherlands": "Europe",

    # North America
    "United States": "North America",
    "USA": "North America",
    "Canada": "North America",
    "Puerto Rico": "North America",
    "Mexico": "North America",

    # South America
    "Brasil": "South America",
    "Brazil": "South America",
    "Argentina": "South America",
    "Columbia": "South America",
    "Colombia": "South America",
    "Chile": "South America",

    # Asia
    "Indonesia": "Asia",
    "South Korea": "Asia",
    "Japan": "Asia",
    "China": "Asia",
    "India": "Asia",

    # Oceania
    "Australia": "Oceania",
    "New Zealand": "Oceania",
}


def parse_country_from_place(place: str) -> str | None:
    """
    founded_place може бути:
    - "Ukraine"
    - "Los Angeles, United States"
    - "Kharkiv, Ukraine"
    Беремо останній шматок після коми як країну.
    """
    if not place or not isinstance(place, str):
        return None
    parts = [p.strip() for p in place.split(",") if p.strip()]
    if not parts:
        return None
    return parts[-1]


# ==== 1. Завантажуємо файли ====
with ULTRA_PATH.open("r", encoding="utf-8") as f:
    ultra_data = json.load(f)

with LOCATIONS_PATH.open("r", encoding="utf-8") as f:
    locations = json.load(f)

parts_of_world = locations["Parts_of_the_World"]

# Для зручності: мапа region_name -> обʼєкт region
region_by_name = {region["name"]: region for region in parts_of_world}

# Допоміжна: повернути список країн у регіоні як set для перевірки
def get_countries_set(region_obj: dict) -> set:
    return {c["name"] for c in region_obj.get("countries", [])}


# ==== 2. Проходимо по всіх піснях та артистах ====
songs = ultra_data["songs"]

for song in songs:
    artist = song.get("artist") or {}
    founded_place = artist.get("founded_place")
    country = parse_country_from_place(founded_place)

    if not country:
        continue

    # Нормалізація простих кейсів
    if country == "USA":
        country = "United States"

    region_name = COUNTRY_TO_REGION.get(country, "unknown")

    # Якщо регіону ще нема в шаблоні — скіпаємо (або можна створити)
    if region_name not in region_by_name:
        # створимо, якщо дуже треба
        region_obj = {"name": region_name, "countries": []}
        parts_of_world.append(region_obj)
        region_by_name[region_name] = region_obj

    region_obj = region_by_name[region_name]

    # у "unknown" в нашій новій структурі теж countries, не cities
    if "countries" not in region_obj:
        region_obj["countries"] = []

    # перевіряємо, чи країна вже є
    existing_countries = get_countries_set(region_obj)

    if country not in existing_countries:
        region_obj["countries"].append({"name": country})


# ==== 3. Зберігаємо оновлений artists_locations ====
with OUTPUT_PATH.open("w", encoding="utf-8") as f:
    json.dump(locations, f, ensure_ascii=False, indent=2)

print(f"✅ Оновлений artists_locations збережено в {OUTPUT_PATH}")
```

```python
import json
from pathlib import Path

INPUT_PATH = Path("artists_locations_updated.json")   # твій оновлений файл
OUTPUT_PATH = Path("artists_locations_countries_only.json")  # куди зберегти

with INPUT_PATH.open("r", encoding="utf-8") as f:
    data = json.load(f)

parts = data.get("Parts_of_the_World", [])

for part in parts:
    name = part.get("name")

    # Якщо це unknown — робимо уніфіковану структуру з пустим списком країн
    if name == "unknown":
        # Якщо в unknown були cities — просто ігноруємо їх
        part["countries"] = []
        # cities на верхньому рівні можна видалити, якщо є
        if "cities" in part:
            del part["cities"]
        continue

    # Для нормальних регіонів гарантуємо наявність "countries"
    countries = part.get("countries", [])
    normalized_countries = []

    for c in countries:
        country_name = c.get("name")
        if not country_name:
            continue
        # Перезаписуємо так, щоб cities завжди був пустим списком
        normalized_countries.append({
            "name": country_name,
            "cities": []
        })

    part["countries"] = normalized_countries

# Зберігаємо результат
with OUTPUT_PATH.open("w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print(f"✅ Збережено оновлений файл без міст у {OUTPUT_PATH}")
```
